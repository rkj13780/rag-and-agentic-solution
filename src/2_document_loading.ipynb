## PDF text extraction
#### PYPDF

```python
import pypdf
from tqdm import tqdm

file_path = "data/f1.pdf"
pdf_data = pypdf.PdfReader(file_path)

pdf_text = ""
for page in tqdm(pdf_data.pages):
    pdf_text += f"{page.extract_text()}\n"

print(pdf_text)
```

#### Read PDF as markdown using pymupdf4llm

```python
import pymupdf4llm
file_path = "data/f1.pdf"
markdown_text = pymupdf4llm.to_markdown(file_path)
```

```python
print(markdown_text)
```
