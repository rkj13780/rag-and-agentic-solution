# Vector database storage

Vector databases use approximate nearest neighbor (ANN) search algorithms.
Indexing methods: 
- FLAT (no index)
- IVF (Inverted file index, uses clustering) 
- LSH (Locality Sensitive Hashing, maximises hash collisions), etc
- HNSW (Hierarchical Navigable Small World)
