```python
import chromadb
```

```python
COLLECTION_NAME = "devday-rag-cosine"

client = chromadb.PersistentClient()
collection = client.get_or_create_collection(COLLECTION_NAME, metadata={"hnsw:space": "cosine"})
```

```python
collection.count()
```

```python
import pymupdf4llm
file_path = "data/f1.pdf"
markdown_text = pymupdf4llm.to_markdown(file_path)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=512,
    chunk_overlap=50,
    length_function=len,
    is_separator_regex=False,
)

recursive_chunks = text_splitter.create_documents([markdown_text])
len(recursive_chunks)
```

```python
collection.add(
    ids=[str(i) for i in range(1, len(recursive_chunks) + 1)],
    documents=[doc.page_content for doc in recursive_chunks]
)
```
